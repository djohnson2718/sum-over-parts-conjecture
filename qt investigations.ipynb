```python
var("t,q,t1,t2")

R.<z> = PowerSeriesRing(SR)
def S(psibar,order):
    return sum((
        z^n*sum((
                prod((t^mu.leg_length(*c) * q^mu.arm_length(*c) 
                     *psibar.subs({x: t^c[0] * q^c[1]}) 
                     /(q^mu.arm_length(*c) - t^(mu.leg_length(*c)+1))
                     /(t^mu.leg_length(*c)-q^(mu.arm_length(*c)+1))
                for c in mu.cells()))
            for mu in Partitions(n)))
    for n in range(order+1))).add_bigoh(order+1)



def pc(f):
    for i in range(f.prec()):
        print (i,"**",f[i].simplify_full().factor())
    
```
