```python
var("t,q,t1,t2,X")

R.<z> = PowerSeriesRing(SR)
def S(psibar,order):
    return sum((
        z^n*sum((
                prod((t^mu.leg_length(*c) * q^mu.arm_length(*c) 
                     *psibar.subs({x: t^c[0] * q^c[1]}) 
                     /(q^mu.arm_length(*c) - t^(mu.leg_length(*c)+1))
                     /(t^mu.leg_length(*c)-q^(mu.arm_length(*c)+1))
                for c in mu.cells()))
            for mu in Partitions(n)))
    for n in range(order+1))).add_bigoh(order+1)



def pc(f):
    for i in range(f.prec()):
        print (i,"**",f[i].simplify_full().factor())
        
def pcb(f):
    fs = qtsubcoef(f)
    for i in range(f.prec()):
        print (i,"**",b(fs[i]).simplify_rational())
        
def Xconst(f):
    return (f/X).residue(X)

def qtsub(f):
    return f.subs({q:exp(t2*X),t:exp(t1*X)})

def qtsubcoef(s):
    return s.map_coefficients(qtsub)

def b(f):
    return Xconst(qtsub(f))
    
```

```python
def S3(order):
    return sum((
        z^n*sum((
                prod(( (- (c[0]*t1 + c[1]*t2))^2
                       / (t1*(mu.leg_length(*c)+1) + t2*(-mu.arm_length(*c)))
                       / (t1*(-mu.leg_length(*c)) + t2^(mu.arm_length(*c)+1))
                for c in mu.cells()))
            for mu in Partitions(n)))
    for n in range(order+1))).add_bigoh(order+1)
```
