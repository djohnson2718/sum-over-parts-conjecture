```python
var("t,q,t1,t2,X")


def S(order, beta):
    return sum((
        z^n*sum(( 
                prod((beta.subs({x: t^mu.leg_length(*c) * q^mu.arm_length(*c)}) for c in mu.cells()))/
                prod((
                    (q^mu.arm_length(*c) - t^(mu.leg_length(*c)+1))*(t^mu.leg_length(*c)-q^(mu.arm_length(*c)+1))
                for c in mu.cells()))
            for mu in Partitions(n)))
    for n in range(order+1))).add_bigoh(order+1)

def Xconst(f):
    return (f/X).residue(X)

def qtsub(f):
    return f.subs({q:exp(t1*X),t:exp(t2*X)})

def b(f):
    return Xconst(qtsub(f))

def check(order,beta):
    logS = S(order,beta).log()
    for i in range(1, order+1):
        print(i,"**",b(logS[i]).simplify_full())

```
